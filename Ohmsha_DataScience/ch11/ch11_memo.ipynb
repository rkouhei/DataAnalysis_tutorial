### 表色系
色を表現するにはさまざまな方法があり、これを表色系と呼ぶ。それぞれの表色系の色要素で構成される空間を色空間と呼ぶ。  
代表的な表色系は次のようなものがある。
- RGB : Red、Green、Blue
- HSV : Hue(色相)、Saturation(彩度)、Value(明度)
- XYZ : Y(輝度)、Z(青み)、X(それ以外)
- CMYK : Cyan、Magenta、Yellow、Key Plate

### エッジ検出
エッジは境界、縁などを意味する。

```python
%matplotlib inline

f_name = '../Sample/data/test.jpg'
img_org = cv2.imread(f_name)
img_gray = cv2.cvtColor(img_org, cv2.COLOR_BGR2GRAY)

edge_sob_x = cv2.Sobel(img_gray,cv2.CV_32F,1,0,ksize=5)
edge_lapl = cv2.Laplacian(img_gray, cv2.CV_32F) 
edge_cann = cv2.Canny(img_gray, 80, 120)

fig = plt.figure()

ax = fig.add_subplot(1, 5, 1)
plt.title('Original Image')
ax.imshow(img_org)
ax = fig.add_subplot(1, 5, 2)
plt.title('Gray scale')
ax.imshow(img_gray)
ax = fig.add_subplot(1, 5, 3)
plt.title('Sobel')
ax.imshow(edge_sob_x)
ax = fig.add_subplot(1, 5, 4)
plt.title('Laplacian')
ax.imshow(edge_lapl)
ax = fig.add_subplot(1, 5, 5)
plt.title('Canny')
ax.imshow(edge_cann)

plt.show()
```
